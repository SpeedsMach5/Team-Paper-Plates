```python
def load_contract():

    # Load the contract ABI
    with open(Path('Eli/abi.json')) as f:
        artwork_abi = json.load(f)

    contract_address = os.getenv("SMART_CONTRACT_ADDRESS")

    # Load the contract
    contract = w3.eth.contract(
        address=contract_address,
        abi=artwork_abi
    )

    return contract
```

```python
contract = load_contract()
```
